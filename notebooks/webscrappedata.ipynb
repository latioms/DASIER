```python
# let's extract data from population.db and make some plots

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

# get data from database
conn = sqlite3.connect('./../population.db')
df = pd.read_sql_query("SELECT * FROM population", conn)
conn.close()

df.head()
```

```python
# show the top 10 countries with the highest population and plotting results
df.sort_values(by='population', ascending=False).head(10)
# plot the top 10 countries with the highest population
plt.figure(figsize=(10, 6))
sns.barplot(x='population', y='country', data=df.sort_values(by='population', ascending=False).head(10))
plt.title('Top 10 countries with the highest population')
plt.show()
```

```python
# Affichons les dix plus grandes densits de population
df.sort_values(by='density', ascending=False).head(10)
# plot the top 10 countries with the highest density
plt.figure(figsize=(10, 6))
sns.barplot(x='density', y='country', data=df.sort_values(by='density', ascending=False).head(10))
plt.title('Top 10 countries with the highest density')
```
